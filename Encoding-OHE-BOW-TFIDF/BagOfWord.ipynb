```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
documents = ["people watch movie and watch movie again",
             "people watch cricket and watch cricket",
             "people like movie and like movie a lot",
             "people like cricket"]
```

```python
bow = CountVectorizer()
```

```python
bow_matrix = bow.fit_transform(documents)
```

```python
print("BoW Matrix:\n", bow_matrix.toarray())
```

```python
print("Vocabulary:", bow.get_feature_names_out())
```

# OOV-issue-ex1

```python
document_1=["Lion is the king of jungle"]
doc1_bow_matrix = bow.transform(document_1)
print("\nDocument_1 BoW matrix:", doc1_bow_matrix.toarray())
```

# OOV-issue-ex2

```python
document_2=["Lion is the king of people jungle"]
doc2_bow_matrix = bow.transform(document_2)
print("\nDocument_2 BoW matrix:\n", doc2_bow_matrix.toarray())
```
