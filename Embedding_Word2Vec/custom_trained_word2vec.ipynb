```python
# data preparation for training word2vec model
folder_path = r"E:\sourcecode\github\dipkvts\ai-python\GenAI-Fullstack\Embedding_Word2Vec\data"
story = []
for filename in os.listdir(folder_path):
  file_path=os.path.join(folder_path,filename)
  with open(file_path,encoding='unicode_escape') as f:
    corpus=f.read()
  raw_sent=sent_tokenize(corpus)
  for sent in raw_sent:
    story.append(simple_preprocess(sent))
```

```python
# list of word-tokens of all sentences of all stories
story
```

```python
# custom word2vec model creation with below specific hyperparameters (variable parameter)
# Hyperparameter: sg=0 for CBOW, sg=1 for skip-gram. If sg not provided it will by default take value 0, ie model will behave like a CBOW model.
custom_model = gensim.models.Word2Vec(
   window=10, # window size
   min_count=5, # any word with total frequency less than this will be ignored
   epochs=5, # number of iterations (forward-backward propagation) over the corpus
   vector_size=150 # dimensionality of word-vectors
  )
```

```python
# total sentence count in corpus
custom_model.corpus_count
```

```python
custom_model.build_vocab(story)
```

```python
custom_model.train(story, total_examples=custom_model.corpus_count, epochs=custom_model.epochs)
```
